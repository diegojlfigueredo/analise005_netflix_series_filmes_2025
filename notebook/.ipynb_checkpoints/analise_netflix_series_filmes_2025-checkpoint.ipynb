```python
# Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Definir o caminho para a pasta de dados
data_path = r'D:\03 - Portfolios\portfolio_diego\analise_netflix_series_filmes_2025\dados'

# Nomes exatos dos arquivos CSV encontrados
movies_file = 'netflix_movies_detailed_up_to_2025.csv'
series_file = 'netflix_tv_shows_detailed_up_to_2025.csv'

# Carregar os datasets
try:
    df_movies = pd.read_csv(os.path.join(data_path, movies_file))
    df_series = pd.read_csv(os.path.join(data_path, series_file))
    print(f"Arquivos '{movies_file}' e '{series_file}' carregados com sucesso!")
except FileNotFoundError:
    print(f"Erro: Um dos arquivos não foi encontrado. Verifique os nomes e o caminho.")
    print(f"Caminho verificado: {data_path}")
    print(f"Arquivos esperados: {movies_file}, {series_file}")
except Exception as e:
    print(f"Ocorreu um erro ao carregar os arquivos CSV: {e}")

# Agora que os DataFrames devem estar carregados, podemos continuar
# Se houver algum erro no carregamento, os próximos passos não serão executados.
if 'df_movies' in locals() and 'df_series' in locals():
    print("\nDataFrames df_movies e df_series estão prontos para análise.")
```

```python
# Função para explorar um DataFrame
def explore_dataframe(df, name):
    print(f"\n--- Exploração Inicial do DataFrame: {name} ---")
    print("\nDimensões do DataFrame (Linhas, Colunas):", df.shape)
    print("\nPrimeiras 5 linhas:")
    print(df.head())
    print("\nInformações sobre as colunas e tipos de dados:")
    print(df.info())
    print("\nEstatísticas Descritivas:")
    print(df.describe(include='all')) # 'all' para incluir colunas não numéricas
    print("\nContagem de valores nulos por coluna:")
    print(df.isnull().sum())
    print("\nPorcentagem de valores nulos por coluna:")
    print((df.isnull().sum() / len(df)) * 100)
    print("-" * 60) # Separador para melhor leitura

# Chamar a função para os DataFrames carregados
if 'df_movies' in locals():
    explore_dataframe(df_movies, "Filmes Netflix")
if 'df_series' in locals():
    explore_dataframe(df_series, "Séries Netflix")
else:
    print("Os DataFrames df_movies ou df_series não foram carregados corretamente. Verifique o Passo 1.")
```

```python
# --- Configurar o caminho para salvar as imagens ---
# Certifique-se de que esta pasta existe ou será criada.
image_path = r'D:\03 - Portfolios\portfolio_diego\analise_netflix_series_filmes_2025\imagens'
os.makedirs(image_path, exist_ok=True) # Cria a pasta se ela não existir

print(f"Pasta para salvar imagens: {image_path}")

# --- Função para Limpeza e Pré-processamento ---
def preprocess_netflix_data(df, name):
    print(f"\n--- Pré-processamento do DataFrame: {name} ---")

    # 1. Tratamento de valores ausentes (NaNs)
    if 'director' in df.columns:
        df.loc[:, 'director'] = df['director'].fillna('Unknown')
    if 'cast' in df.columns:
        df.loc[:, 'cast'] = df['cast'].fillna('Unknown')

    if 'country' in df.columns:
        df.loc[:, 'country'] = df['country'].fillna('Unknown')

    if 'date_added' in df.columns:
        df.loc[:, 'date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

        # --- DEBUG PRINTS AQUI ---
        print(f"DEBUG: Dtype de 'date_added' após conversão para {name}: {df['date_added'].dtype}")
        print(f"DEBUG: Primeiras 5 datas adicionadas de 'date_added' para {name}:")
        print(df['date_added'].head())
        print(f"DEBUG: Contagem de nulos em 'date_added' para {name}: {df['date_added'].isnull().sum()}")
        # -------------------------

        # Se houver NaNs após a conversão, preencher com a data de adição mais frequente
        if df['date_added'].isnull().any():
            try:
                # Preenche com a moda se houver datas válidas
                df.loc[:, 'date_added'].fillna(df['date_added'].mode()[0], inplace=True)
            except IndexError: # Se mode() retornar vazio (ex: todos NaT ou todos nulos)
                print(f"Aviso: Coluna 'date_added' em {name} contém apenas valores nulos ou inválidos. Preenchendo com '2000-01-01'.")
                df.loc[:, 'date_added'] = df['date_added'].fillna(pd.to_datetime('2000-01-01'))


    if 'rating' in df.columns:
        df.loc[:, 'rating'] = df['rating'].fillna(df['rating'].mode()[0])

    # 2. Conversão de Tipos de Dados e Criação de Novas Colunas
    if 'date_added' in df.columns and not df['date_added'].isnull().all():
        # Verifique se a coluna date_added é realmente datetime antes de usar .dt
        if pd.api.types.is_datetime64_any_dtype(df['date_added']):
            df.loc[:, 'month_added'] = df['date_added'].dt.month
            df.loc[:, 'year_added'] = df['date_added'].dt.year
            df.loc[:, 'year_added'] = df['year_added'].fillna(0).astype(int)
        else:
            print(f"Aviso: 'date_added' em {name} não é datetime. Pulando extração de mês/ano.")


    if 'duration' in df.columns:
        if 'Movie' in name: # Para o DataFrame de filmes
            df.loc[:, 'duration_min'] = df['duration'].str.replace(' min', '', regex=False).astype(float)
            df.loc[:, 'duration_min'] = df['duration_min'].fillna(df['duration_min'].mean()).astype(int)
        elif 'Séries' in name: # Para o DataFrame de séries
            df.loc[:, 'num_seasons'] = df['duration'].str.replace(' Season(s)?', '', regex=True).astype(float)
            df.loc[:, 'num_seasons'] = df['num_seasons'].fillna(df['num_seasons'].mode()[0]).astype(int)

    # 3. Tratamento de colunas de lista (ex: 'genres', 'country', 'cast', 'director')
    if 'genres' in df.columns:
        df.loc[:, 'genres'] = df['genres'].apply(lambda x: [i.strip() for i in str(x).split(',')])

    if 'country' in df.columns:
        df.loc[:, 'country'] = df['country'].apply(lambda x: [i.strip() for i in str(x).split(',')])

    if 'cast' in df.columns:
        df.loc[:, 'cast'] = df['cast'].apply(lambda x: [i.strip() for i in str(x).split(',')])

    if 'director' in df.columns:
        df.loc[:, 'director'] = df['director'].apply(lambda x: [i.strip() for i in str(x).split(',')])

    if 'Movie' in name:
        if 'budget' in df.columns:
            df.loc[:, 'budget'] = df['budget'].fillna(0)
        if 'revenue' in df.columns:
            df.loc[:, 'revenue'] = df['revenue'].fillna(0)

    print(f"Pré-processamento de '{name}' concluído.")
    print("Novas 5 linhas após pré-processamento:")
    print(df.head())
    print("-" * 60)
    return df

# Aplicar a função de pré-processamento aos DataFrames
if 'df_movies' in locals():
    df_movies = preprocess_netflix_data(df_movies.copy(), "Filmes Netflix")
if 'df_series' in locals():
    df_series = preprocess_netflix_data(df_series.copy(), "Séries Netflix")
```

```python
# Certifique-se de que as bibliotecas e o image_path estão definidos e a pasta existe
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from collections import Counter # Já importado, mas reforçando

# Assumindo que image_path já está definido do código anterior
# image_path = r'D:\03 - Portfolios\portfolio_diego\analise_netflix_series_filmes_2025\imagens'
# os.makedirs(image_path, exist_ok=True) # Criar a pasta se não existir

sns.set_style("whitegrid")
plt.rcParams['figure.dpi'] = 100

# --- Análise 4: Distribuição de Classificações (Ratings) ---

# Unir os DataFrames para uma visão geral da distribuição de ratings, se desejar.
# Ou analisar separadamente para filmes e séries. Vamos analisar separadamente.

# Rating de Filmes
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=df_movies, order=df_movies['rating'].value_counts().index, palette='crest')
plt.title('Distribuição de Classificações (Ratings) para Filmes na Netflix')
plt.xlabel('Classificação')
plt.ylabel('Número de Filmes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(image_path, 'distribuicao_ratings_filmes.png'))
plt.show()

# Rating de Séries
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=df_series, order=df_series['rating'].value_counts().index, palette='flare')
plt.title('Distribuição de Classificações (Ratings) para Séries na Netflix')
plt.xlabel('Classificação')
plt.ylabel('Número de Séries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(image_path, 'distribuicao_ratings_series.png'))
plt.show()

# --- Análise 5: Top Países Produtores de Conteúdo ---

# Função auxiliar para pegar os top N de uma coluna com múltiplos valores (países)
def get_top_n_countries(df, n=10):
    all_countries = [country for sublist in df['country'] for country in sublist if country != 'Unknown']
    country_counts = Counter(all_countries)
    return pd.DataFrame(country_counts.most_common(n), columns=['Country', 'Count'])

# Top Países Produtores de Filmes
top_movie_countries = get_top_n_countries(df_movies, 10)
plt.figure(figsize=(12, 7))
sns.barplot(x='Count', y='Country', data=top_movie_countries, palette='rocket')
plt.title('Top 10 Países Produtores de Filmes na Netflix')
plt.xlabel('Número de Filmes')
plt.ylabel('País')
plt.tight_layout()
plt.savefig(os.path.join(image_path, 'top_10_paises_filmes.png'))
plt.show()

# Top Países Produtores de Séries
top_series_countries = get_top_n_countries(df_series, 10)
plt.figure(figsize=(12, 7))
sns.barplot(x='Count', y='Country', data=top_series_countries, palette='mako')
plt.title('Top 10 Países Produtores de Séries na Netflix')
plt.xlabel('Número de Séries')
plt.ylabel('País')
plt.tight_layout()
plt.savefig(os.path.join(image_path, 'top_10_paises_series.png'))
plt.show()

# --- Análise 6: Duração de Filmes e Número de Temporadas de Séries ---

# Duração de Filmes (duration_min)
if 'duration_min' in df_movies.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df_movies['duration_min'], bins=30, kde=True, color='purple')
    plt.title('Distribuição da Duração de Filmes na Netflix (Minutos)')
    plt.xlabel('Duração (Minutos)')
    plt.ylabel('Número de Filmes')
    plt.tight_layout()
    plt.savefig(os.path.join(image_path, 'distribuicao_duracao_filmes.png'))
    plt.show()
else:
    print("A coluna 'duration_min' não foi encontrada para filmes. Verifique o pré-processamento.")

# Número de Temporadas de Séries (num_seasons)
if 'num_seasons' in df_series.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='num_seasons', data=df_series, order=df_series['num_seasons'].value_counts().index, palette='viridis')
    plt.title('Distribuição do Número de Temporadas de Séries na Netflix')
    plt.xlabel('Número de Temporadas')
    plt.ylabel('Número de Séries')
    plt.tight_layout()
    plt.savefig(os.path.join(image_path, 'distribuicao_num_temporadas_series.png'))
    plt.show()
else:
    print("A coluna 'num_seasons' não foi encontrada para séries. Verifique o pré-processamento.")

print("\nAnálise Exploratória de Dados (EDA) finalizada. Todos os gráficos devem estar salvos na pasta 'imagens'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- 1. Carregar os DataFrames e Pré-processamento ---

# Carregar os conjuntos de dados com o caminho corrigido
try:
    df_movies = pd.read_csv('../dados/netflix_movies_detailed_up_to_2025.csv')
    df_shows = pd.read_csv('../dados/netflix_tv_shows_detailed_up_to_2025.csv')
    print("DataFrames df_movies e df_shows carregados com sucesso!")
except FileNotFoundError:
    print("Erro: Arquivos CSV não encontrados! Verifique se estão em '../dados/'.")
    # Se os arquivos ainda não forem encontrados, a execução parará aqui.
    exit() # Adicionado para evitar que o código continue e gere mais erros se os arquivos não forem encontrados

# Pré-processamento: Converter 'date_added' para datetime
df_movies['date_added'] = pd.to_datetime(df_movies['date_added'], errors='coerce')
df_shows['date_added'] = pd.to_datetime(df_shows['date_added'], errors='coerce')

# Pré-processamento: Preencher valores NaN em colunas importantes
# Usando a sintaxe recomendada pelo FutureWarning para evitar o aviso.
df_movies['director'] = df_movies['director'].fillna('Unknown')
df_shows['director'] = df_shows['director'].fillna('Unknown')
df_movies['cast'] = df_movies['cast'].fillna('Unknown')
df_shows['cast'] = df_shows['cast'].fillna('Unknown')
df_movies['country'] = df_movies['country'].fillna('Unknown')
df_shows['country'] = df_shows['country'].fillna('Unknown')

print("Pré-processamento básico de filmes e séries concluído.")

# Criar a pasta 'imagens' se ela não existir
if not os.path.exists('imagens'):
    os.makedirs('imagens')

# --- 2. Funções Auxiliares (para limpeza e explosão de colunas) ---

def clean_genres_column(df):
    df_temp = df.copy()
    if 'genres' in df_temp.columns:
        # CORREÇÃO AQUI: Converter para string e preencher NaN antes de splitar
        df_temp['genres'] = df_temp['genres'].astype(str).fillna('') # Garante que é string e trata NaNs
        df_temp['genres'] = df_temp['genres'].apply(lambda x: [s.strip() for s in x.split(',') if s.strip() != '']) # Evita strings vazias resultantes de NaNs
    else:
        print(f"Aviso: Coluna 'genres' não encontrada no DataFrame.")
        df_temp['genres'] = [[]] * len(df_temp) # Cria uma coluna vazia para evitar erro
    return df_temp

def clean_and_explode_column(df, column_name):
    df_temp = df.copy()
    df_temp[column_name] = df_temp[column_name].fillna('Unknown').astype(str) # Garante que é string e trata NaNs
    df_temp[column_name] = df_temp[column_name].apply(lambda x: [s.strip() for s in str(x).split(',')])
    return df_temp.explode(column_name)

# --- 3. Geração dos Gráficos ---

print("\nGerando gráficos...")

# 3.1. Gráficos de Conteúdo por Ano de Adição e Lançamento
# Certifique-se de que a coluna 'year_added' e 'release_year' estão disponíveis
df_movies['year_added'] = df_movies['date_added'].dt.year
df_shows['year_added'] = df_shows['date_added'].dt.year

# Distribuição de Conteúdo por Ano de Adição
plt.figure(figsize=(14, 7))
sns.histplot(df_movies['year_added'].dropna(), bins=30, kde=False, color='skyblue', label='Filmes')
sns.histplot(df_shows['year_added'].dropna(), bins=30, kde=False, color='lightcoral', label='Séries')
plt.title('Distribuição de Conteúdo por Ano de Adição à Plataforma')
plt.xlabel('Ano de Adição')
plt.ylabel('Número de Títulos')
plt.legend()
plt.tight_layout()
plt.savefig('imagens/distribuicao_conteudo_por_ano_adicao.png')
plt.close()

# Distribuição de Conteúdo por Ano de Lançamento
plt.figure(figsize=(14, 7))
sns.histplot(df_movies['release_year'].dropna(), bins=30, kde=False, color='skyblue', label='Filmes')
sns.histplot(df_shows['release_year'].dropna(), bins=30, kde=False, color='lightcoral', label='Séries')
plt.title('Distribuição de Conteúdo por Ano de Lançamento')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Número de Títulos')
plt.legend()
plt.tight_layout()
plt.savefig('imagens/distribuicao_conteudo_por_ano_lancamento.png')
plt.close()

# 3.2. Top Gêneros (Filmes e Séries)
df_movies_genres = clean_genres_column(df_movies)
df_shows_genres = clean_genres_column(df_shows)
genres_movies_exploded = df_movies_genres.explode('genres')
genres_shows_exploded = df_shows_genres.explode('genres')
top_10_genres_movies = genres_movies_exploded['genres'].value_counts().head(10)
top_10_genres_shows = genres_shows_exploded['genres'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_genres_movies.values, y=top_10_genres_movies.index, palette='viridis')
plt.title('Top 10 Gêneros de Filmes na Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('Gênero')
plt.tight_layout()
plt.savefig('imagens/top_10_generos_filmes.png')
plt.close()

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_genres_shows.values, y=top_10_genres_shows.index, palette='magma')
plt.title('Top 10 Gêneros de Séries na Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('Gênero')
plt.tight_layout()
plt.savefig('imagens/top_10_generos_series.png')
plt.close()
print("Gráficos de Top Gêneros gerados e salvos.")

# 3.3. Top Diretores (Filmes e Séries)
directors_movies_exploded = clean_and_explode_column(df_movies, 'director')
directors_shows_exploded = clean_and_explode_column(df_shows, 'director')
top_10_directors_movies = directors_movies_exploded['director'].value_counts().drop('Unknown', errors='ignore').head(10)
top_10_directors_shows = directors_shows_exploded['director'].value_counts().drop('Unknown', errors='ignore').head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_directors_movies.values, y=top_10_directors_movies.index, palette='cubehelix')
plt.title('Top 10 Diretores de Filmes na Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('Diretor')
plt.tight_layout()
plt.savefig('imagens/top_10_diretores_filmes.png')
plt.close()

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_directors_shows.values, y=top_10_directors_shows.index, palette='mako')
plt.title('Top 10 Diretores de Séries na Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('Diretor')
plt.tight_layout()
plt.savefig('imagens/top_10_diretores_series.png')
plt.close()
print("Gráficos de Top Diretores gerados e salvos.")

# 3.4. Top Atores (Filmes e Séries)
cast_movies_exploded = clean_and_explode_column(df_movies, 'cast')
cast_shows_exploded = clean_and_explode_column(df_shows, 'cast')
top_10_cast_movies = cast_movies_exploded['cast'].value_counts().drop('Unknown', errors='ignore').head(10)
top_10_cast_shows = cast_shows_exploded['cast'].value_counts().drop('Unknown', errors='ignore').head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cast_movies.values, y=top_10_cast_movies.index, palette='rocket')
plt.title('Top 10 Atores/Atrizes de Filmes na Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('Ator/Atriz')
plt.tight_layout()
plt.savefig('imagens/top_10_atores_filmes.png')
plt.close()

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cast_shows.values, y=top_10_cast_shows.index, palette='crest')
plt.title('Top 10 Atores/Atrizes de Séries na Netflix')
plt.xlabel('Número de Títulos')
plt.ylabel('Ator/Atriz')
plt.tight_layout()
plt.savefig('imagens/top_10_atores_series.png')
plt.close()
print("Gráficos de Top Atores/Atrizes gerados e salvos.")

# 3.5. Distribuição do Número de Temporadas de Séries
# Correção aqui: garantir que 'duration' (que agora é a coluna 'num_seasons') seja numérica
# e lidar com o nome da coluna no df_shows
if 'duration' in df_shows.columns:
    df_shows['num_seasons'] = df_shows['duration'].str.extract(r'(\d+)').astype(float)
    plt.figure(figsize=(10, 6))
    sns.countplot(x='num_seasons', data=df_shows.dropna(subset=['num_seasons']), palette='viridis')
    plt.title('Distribuição do Número de Temporadas de Séries na Netflix')
    plt.xlabel('Número de Temporadas')
    plt.ylabel('Número de Séries')
    plt.tight_layout()
    plt.savefig('imagens/distribuicao_num_temporadas_series.png')
    plt.close()
    print("Gráfico de Distribuição do Número de Temporadas de Séries gerado e salvo.")
else:
    print("Aviso: Coluna 'duration' (para número de temporadas) não encontrada em df_shows. Gráfico não gerado.")

print("\nTodos os gráficos necessários foram gerados e salvos na pasta 'imagens/'.")
```
